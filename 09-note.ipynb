問題106

```python
import random
while True:
    rand_num = random.randint(1,6)
    print(rand_num)
    if rand_num == 4:
        break
    print('.')
```

breakの後にprint('.')にしているため4が出ると.は表示されない

```python
import random
while True:
    rand_num = random.randint(1,6)
    print(rand_num)
    if rand_num == 4:
        continue
    print('.')
```
